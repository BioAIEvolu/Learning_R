## R的基本数据类型
- R中最基本对的单位就是向量，单个元素就是长度为1的向量
- 创建向量的方法是c()

```r
### 因子型
```

```r
sex <- c('F','M', 'F', 'M', 'F')
sexf <- factor(sex)
sexf
```

### 数值型

```r
num1 <- c(9.1, 8.2, 7.3)
num2 <- 1:10
```

### 字符型

```r
char <- c('a', "b", 'c')
char
```

### 布尔逻辑型

```r
logit <- c(TRUE,FALSE,T,F)
logit
```

```r
# 只要有字符型 所有元素都会自动被转化为字符型的向量
c(1, 'a', TRUE)
```

### 数值操作

```r
# 切片、赋值操作
num1[1] <- 100
num1[c(1, 3)] <- c(99.9, 77.7)
num1[c(T,F,T)]
num1[-2]
```

```r
#在向量中给元素命名
num3 <- c('First'=9.1, 'second'=8.2, 'third'=7.3)
num3[c('First', 'third')]
```

```r
#通过表达式得到的布尔值向量进行索引
a <- 20:30
a%%2 == 0
a[a%%2 == 0]  # 保留偶数
```

```r
# 查看数据类型的方法class()
class(num1)
class(char)
class(logit)
class(sexf)
```

```r
is.character(num1)  #返回布尔值
#类型转换
num1 = as.character(num1)
class(num1)
```

```r
#更多例子和推广
is.character(char)
is.character(logit)
as.character(logit)
```

## 复杂数据类型

### 矩阵
所有的元素类型都需要相同

```r
m <- matrix(1:6, nrow=2, ncol=3, byrow=T, dimnames=list(c('r1', 'r2'), c('c1', 'c2', 'c3')))
m[1,1] #选取元素
m[1,] #选取第一行
m[,1] #全取第一列
m[c(T, F),] #第一行
m['r1',] #第一行
```

```r
#修改元素
m[1,] <- c(100,99,98)
m[1,] <- 100
t(m)
```

### 数组
数组维度可以大于2

```r
array(1:10)
```

```r
array(1:10, dim=c(2,5))
```

```r
dim1 <- c('A1', 'A2')
dim2 <- c('B1', 'B2', 'B3')
dim3 <- c('C1', 'C2')
arr3 <- array(1:12, dim=c(2,3,2), dimnames=list(dim1, dim2, dim3))   
```

```r
#选取元素
arr3[,,'C1']
arr3[,,1]
arr3[,,c(T,F)]
```

### 数据框
data.frame()

```r
student <- data.frame(ID=c(1001, 1002, 1003),
                      NAME=c('lili', 'bob', 'sam'),
                      GENDER=c('F','M','M'),
                      stringsAsFactors = F)
```

```r
student[1,1] <- 1009
student[1,]
student[,1]
student[,'ID']  #返回向量
student['ID']   #返回数据框
student$ID      #返回向量
student[c('ID', 'NAME')]
```
